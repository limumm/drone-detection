```python

def find_different_indexes(data1, data2):
    different_indexes = []
    for i in range(len(data1['res'])):
        if data1['res'][i] != data2['res'][i]:
            different_indexes.append(i)
    return different_indexes

different_indexes = find_different_indexes(data1, data2)
print(f"Indexes where data1 and data2 are different: {different_indexes}")
```
